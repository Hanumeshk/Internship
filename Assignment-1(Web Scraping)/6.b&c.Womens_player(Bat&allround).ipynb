# Importing Libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
from urllib.request import urlopen
import numpy as np
```

```python
url='https://www.icc-cricket.com/rankings/womens/player-rankings/odi'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
other_position=[]
for i in soup.find_all('span',class_="rankings-table__pos-number"):
    other_position.append(i.text)
other_position
```

```python
positions=pd.DataFrame(other_position,columns=['position'])
positions.loc[0:8]
```

```python
other_pos=positions['position'].str.split('\n',expand=True)
posi=other_pos.loc[0:8,1:1]
posi
```

```python
team=[]
for i in soup.find_all('span',class_="table-body__logo-text"):
    team.append(i.text)
team
```

```python
teams=pd.DataFrame(team)
team=teams.loc[0:8]
team
```

```python
other_bat=[]
for i in soup.find_all('td',class_="table-body__cell name"):
    other_bat.append(i.text)

other_bat
```

```python
bat=pd.DataFrame(other_bat,columns=['Player'])
bat
```

```python
Batsman=bat['Player'].str.split('\n',expand=True)
other_batsman=Batsman.loc[0:8,1:1]
other_batsman
```

```python
rating=[]
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
rating
```

```python
other_rating=pd.DataFrame(rating)
ratings=other_rating.loc[0:8]
ratings
```

```python
pos=pd.concat([posi,other_batsman,team,ratings],axis=1)
pos
```
