# 5) b) Top 10 ODI Batsmen along with the records of their team and rating.

# Importing Libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
from urllib.request import urlopen
```

```python
data=urlopen('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
data
```

```python
soup=BeautifulSoup(data,"html.parser")
soup
```

```python

first_data=[]
for i in soup.find_all('div',class_="rankings-block__top-player"):
    first_data.append(i.text)
first_data
```

```python
df=pd.DataFrame(first_data,columns=['top'])
df
```

```python
df2=df['top'].str.split('\n',expand=True)
df3=df2.loc[0:2,4:17]
df3
```

```python
first_top=df3.loc[0:0,[4,13,16,17]]
first_top.columns=["Position","Player","Team","Rating"]
first_top
```

```python
position=[]
for i in soup.find_all('span',class_="rankings-table__pos-number"):
    position.append(i.text)
position
```

```python
icc_pos=pd.DataFrame(position,columns=['position'])
icc_posi=icc_pos['position'].str.split('\n',expand=True)
icc_posit=icc_posi.loc[0:8,1:1]
icc_posit.columns=["Position"]
icc_position=icc_posit
icc_position
```

```python
player=[]
for i in soup.find_all('td',class_="table-body__cell name"):
    player.append(i.text)
player
```

```python
icc_player=pd.DataFrame(player,columns=['player'])
icc_bat=icc_player['player'].str.split('\n',expand=True)
icc_batting=icc_bat.loc[0:8,1:1]
icc_batting.columns=["Player"]
icc_batting
```

```python
team=[]
for i in soup.find_all('span',class_="table-body__logo-text"):
    team.append(i.text)
team
```

```python
icc_team=pd.DataFrame(team,columns=['Team'])
icc_teams=icc_team.loc[0:8]
icc_teams
```

```python
rating=[]
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
rating
```

```python
icc_rating=pd.DataFrame(rating,columns=['Rating'])
icc_ratings=icc_rating.loc[0:8]
icc_ratings
```

```python
df=pd.concat([icc_position,icc_batting,icc_teams,icc_ratings],axis=1)
df
```

# B.World Top 10 Batsmans

```python
Top_batsman=pd.concat([first_top,df],axis=0,ignore_index=True)
Top_batsman
```
