# Importing Libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
```

```python
url='https://presidentofindia.nic.in/former-presidents.htm'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
names=[]

for i in soup.find_all('div',class_="presidentListing"):
    names.append(i.text)
    
names
```

```python
Presi=pd.DataFrame(names,columns=['name'])
President=Presi['name'].str.split('\n',expand=True)
Ind_pres=President.loc[:,1:2]
Ind_pres.columns=["Names","Term of Office"]
Ind_pres
```
