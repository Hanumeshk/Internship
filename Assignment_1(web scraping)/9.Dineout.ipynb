# 9)Write a python program to scrape mentioned details from dineout.co.in :


i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import requests
import re
import json
```

```python
url='https://www.dineout.co.in/bangalore-restaurants/welcome-back'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
rest=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    rest.append(i.text)
rest
```

```python
df=pd.DataFrame(rest,columns=['Restaurant'])
df
```

```python
loc=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
loc
```

```python
df1=pd.DataFrame(loc,columns=["Location"])
df1
```

```python
cuis=[]
for i in soup.find_all('div',class_="detail-info"):
    cuis.append(i.text)
cuis
```

```python
df2=pd.DataFrame(cuis,columns=['cuis'])
df3=df2['cuis'].str.split('|',expand=True)
df4=df3.loc[:,1:1]
df4.columns=["Location"]
df4
```

```python
rate=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rate.append(i.text)
rate
```

```python
rating=pd.DataFrame(rate,columns=['Rating'])
rating
```

```python
Dineout=pd.concat([df,df1,df4,rating],axis=1)
Dineout
```

```python
img=[]
link=soup.find_all('div',class_="img-wrap")
for i in link:
    j=i.img["data-src"]
    img.append(j)
    

img
```

```python
img_link=pd.DataFrame(img,columns=['Image_link'])
img_link
```

```python
Dineout_Rest=pd.concat([Dineout,img_link],axis=1)
Dineout_Rest
```
