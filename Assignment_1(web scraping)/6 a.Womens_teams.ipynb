# 6 a) Top 10 ODI teams in womenâ€™s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
from urllib.request import urlopen
```

```python
url='https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
position=[]
for i in soup.find_all('tr',class_="rankings-block__banner"):
    position.append(i.text)
position    
```

```python
other_team=[]
for i in soup.find_all('tr',class_="table-body"):
    other_team.append(i.text)
other_team
```

```python
other=pd.DataFrame(other_team,columns=['position'])
other
```

```python
others=other['position'].str.split('\n',expand=True)
others
```

```python
other_tea=others.loc[:,[1,4,7,8,9]]
other_tea
```

```python
other_tea.columns=['position','team','matches','points','ratings']
df4=other_tea
df4
```

```python
df1=pd.DataFrame(position,columns=['position'])
df1
```

```python
df=df1['position'].str.split('\n',expand=True)
df
```

```python
df2=df.loc[:,[1,4,7,8,10]]
df2
```

```python
df2.columns=['position','team','matches','points','ratings']
df3=df2
df3
```

```python
Womens=pd.concat([df3,df4])
Womens=Womens.drop(9)
Womens
```
